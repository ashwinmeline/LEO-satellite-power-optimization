```python
import sys
import os

root_path = os.path.abspath(os.path.join(os.getcwd(), '..'))
if root_path not in sys.path:
    sys.path.append(root_path)

import numpy as np
import matplotlib.pyplot as plt
from src.physics import compute_eclipse_duration
from src.power import calculate_required_capacity

print("Modules imported successfully!")
```

```python
# Setup the trade study range
altitudes = np.linspace(400, 800, 50)
eclipses = compute_eclipse_duration(altitudes)
capacities = calculate_required_capacity(eclipses, p_payload_w=50)

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 5))
ax1.plot(altitudes, eclipses/60, color='tab:blue', label='Eclipse Duration')
ax1.set_xlabel('Altitude (km)')
ax1.set_ylabel('Eclipse Time (min)', color='tab:blue')

ax2 = ax1.twinx()
ax2.plot(altitudes, capacities, color='tab:red', linestyle='--', label='Battery Capacity')
ax2.set_ylabel('Required Capacity (Wh)', color='tab:red')

plt.title('LEO Constellation Power Trade Study')
plt.grid(True, alpha=0.3)
plt.show()
```
